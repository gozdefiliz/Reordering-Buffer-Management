```python
#all necessary side algorithms

#finds the number of color changes in a given list
def nb_of_change(x):
    count=0
    i=0
    while i<(len(x)-1):
        if x[i]!=x[i+1]:
            count=count+1
            i=i+1
        else:
            i=i+1
    return count

#adds k items to the buffer from the input
def add(k, data,buffer):
    while k>0:
        buffer.append(data.pop(0))
        k=k-1
        
#finds and lists the unique items in a given list
def unique(list1): 
    unique_list = [] 
    for x in list1: 
        if x not in unique_list: 
            unique_list.append(x)
    return unique_list

#finds the most freq color in buffer
def most_freq(List): 
    return max(set(List), key = List.count)

#evicts c_act, appends output, removes from buffer, fills buffer till there is no c_act in buffer 
def evict(c_act, data, buffer,output_,k):
    while (c_act in buffer)==True:
        count=0
        buffer.remove(c_act)
        count=count+1
        for i in range(0,count):
            output_.append(c_act)
        while len(buffer)<k and len(data)!=0:
            add(k-len(buffer), data,buffer)
    return output_

#finds the second most frequent color in a given list
def second_most_freq(List): 
    from collections import Counter 
    occurence_count = Counter(List) 
    if len(occurence_count)>=2:
        return occurence_count.most_common(2)[1][0]
    else:
        return -1
    
#finds the occurence of a given color in a given list      
def occurence(List, x): 
    count=0
    for i in range (0, len(List)):
        if List[i]==x:
            count=count+1
    return count


#finds the least freq color in buffer
def least_freq(List): 
    return min(set(List), key = List.count)
```

```python
def RCChooseNewActiveColor(buffer):
    import random
    c_active=random.choice(buffer)
    return c_active

def RC(k,input_):
    buffer=input_[0:0]
    output_=input_[0:0]
    data=input_[:]
    x=nb_of_change(data)
    add(k,data,buffer)
    result=[0]*1
    c_active=-1
    while len(buffer)!=0:
        if c_active not in buffer:
            c_active=RCChooseNewActiveColor(buffer)
            evict(c_active,data, buffer,output_,k)
    result[0]=float(nb_of_change(output_)/x)
    print(result)
```
