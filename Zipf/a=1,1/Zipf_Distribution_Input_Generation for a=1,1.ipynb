```python
#truncated zipf, to have color size of 5 max_value needs to be 5
def truncated_Zipf(a, max_value, size):
    import scipy.stats as sct
    temp_size = size
    while True:
        temp_size *= 2
        temp = sct.zipf.rvs(a, size=temp_size)
        truncated = temp[temp <= max_value]
        if len(truncated) >= size:
            return truncated[:size]
```
