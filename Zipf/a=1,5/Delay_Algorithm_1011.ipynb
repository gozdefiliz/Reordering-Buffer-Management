```python
#all necessary side algorithms

#finds the number of color changes in a given list
def nb_of_change(x):
    count=0
    i=0
    while i<(len(x)-1):
        if x[i]!=x[i+1]:
            count=count+1
            i=i+1
        else:
            i=i+1
    return count

#adds k items to the buffer from the input
def add(k, data,buffer):
    while k>0:
        buffer.append(data.pop(0))
        k=k-1
        
#finds and lists the unique items in a given list
def unique(list1): 
    unique_list = [] 
    for x in list1: 
        if x not in unique_list: 
            unique_list.append(x)
    return unique_list

#finds the most freq color in buffer
def most_freq(List): 
    return max(set(List), key = List.count)

#evicts c_act, appends output, removes from buffer, fills buffer till there is no c_act in buffer 
def evict(c_act, data, buffer,output_,k):
    while (c_act in buffer)==True:
        count=0
        buffer.remove(c_act)
        count=count+1
        for i in range(0,count):
            output_.append(c_act)
        while len(buffer)<k and len(data)!=0:
            add(k-len(buffer), data,buffer)
    return output_

#finds the second most frequent color in a given list
def second_most_freq(List): 
    from collections import Counter 
    occurence_count = Counter(List) 
    if len(occurence_count)>=2:
        return occurence_count.most_common(2)[1][0]
    else:
        return -1
    
#finds the occurence of a given color in a given list      
def occurence(List, x): 
    count=0
    for i in range (0, len(List)):
        if List[i]==x:
            count=count+1
    return count


#finds the least freq color in buffer
def least_freq(List): 
    return min(set(List), key = List.count)
```

```python
def delay(k,input_):
    import numpy as np
    import math
    buffer=input_[0:0]
    output_=input_[0:0]
    data=input_[:]
    nb_of_delays=0
    c_act=-1
    result=[0]*1
    
   #data tükenmemişken 
    while len(data)!=0:
        ex=0
        while len(buffer)<k and len(data)!=0:
            #buffera k-len(buffer) eleman ekle
            add(min((k-len(buffer)), len(data)), data, buffer)
            
            #eğer yeni eklenen elemanlar arasında bir önceki aşamada evict edilmiş eleman varsa onları çıkart
            if (c_act!=-1 and (c_act in buffer)==True):
                while (c_act in buffer)==True:
                    buffer.remove(c_act)
                    output_.append(c_act)
                    #her çıkardığın eleman için buffera 1 eleman ekle
                    add(min(1,len(data)), data, buffer)
            else:
                continue
            if len(buffer)==0:
                break
                
            #nb of colors in the buffer is sigma
            sigma=len(unique(buffer))
            #most frequent color in the buffer
            o_1=most_freq(buffer)
            size_of_o1=occurence(buffer,o_1)
        
            #1.kurala uygunluğunu dene
            while (size_of_o1 >= 2 * (math.ceil(len(buffer)/sigma))) and ex<len(data):
                #kural bozulduğu sürece sigmayı düşürecek hareketi yapalım-en az varlığı olan rengi tamamen bufferdan 
                #çıkarıp datanın sonuna atalım
                delayed=least_freq(buffer)
                while (delayed in buffer)==True:
                    nb_of_delays=nb_of_delays+1
                    ex=ex+1
                    buffer.remove(delayed)
                    data.append(delayed)
                if len(buffer)==0:
                    break
                    
                    
        #bu whiledan çıktığımda 1.buffer size k ya ulaştı ve kural tutuyor, 2. datada eleman kalmadı, 3.datada kalan her elemanı
        #delay etmek zorunda kalıyor
        
        #her sonuçta(eğer buffer boş değilse) bufferdan most frequent elemanı çıkarıp outputa eklemeliyim
        if len(buffer)!=0:
            c_act=most_freq(buffer)
            while(c_act in buffer)==True:
                buffer.remove(c_act)
                output_.append(c_act)
            
        #bunu yaptıktan sonra 1.while'a geri dönüp datada eleman olduğu sürece aynı işi yapacağım
    
    #1.whiledan datada eleman bittiği zaman çıkacağım, çıktığımda bufferda eleman kalmış olabilir
    while len(buffer)!=0:
        c_act=most_freq(buffer)
        while (c_act in buffer)==True:
            buffer.remove(c_act)
            output_.append(c_act)
            
    result[0]=float(nb_of_change(output_)/nb_of_change(input_))
    print(result)

```
